```python
#LINEAR REGRESSION
#Linear regression is supervised ML algorithm that predict the continuos values.
#Eg: house price, salary prediction, marks prediction

#Formula: y=mx+c

#for multiple feature: y=w1x1+w2x2+...+wnxn+b

#why LR is important
#1.This is basic building block of ML
#2.Easy to understand how real ML model learn

#What is model work
#Model weights set that the prediction error is less to less.

#Error measure 
#MSE = 1/n∑(ytrue​−ypred​)2

```

```python
#Basic Linear Regreesion(Single Feature)
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Create a dummy dataset
#x = input feature
X = np.array([1,2,3,4,5,6]).reshape(-1,1)

#y = target
y = np.array([3,4,2,4,5,7])
```

```python
#Model Train
model = LinearRegression()
model.fit(X, y)
```

```python
#Predict
pred = model.predict(X)
```

```python
#Plot
plt.scatter(X,y)
plt.plot(X,pred)
plt.show()
```

```python
print("Slope (m):",model.coef_)
print("Intercept (c):", model.intercept_)
```

```python
#Multiple Linear Regression
#Housing example-- area, bedrooms--price
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Dataset
data = {
    'area':[1000, 1500, 1800, 2400, 300],
    'bedrooms':[2,3,3,4,5],
    'price':[50, 65, 75, 90, 120]
}
df = pd.DataFrame(data)
```

```python
#Split
X = df[['area','bedrooms']]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Train
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#Pwefoemance Metrices
from sklearn.metrics import mean_squared_error, r2_score
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
```
