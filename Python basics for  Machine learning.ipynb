```python
#Python basics for ML
#Variable and Data Types
# Variables are used to store the data

name = "Abhinadan"
age = 24
height = 5.8
is_student = True

print("Name:", name)
print("age:", age)
print("height:", height)
print("student:", is_student)

print("Data types")
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```

```python
# List - Store multiple values
#List are Ordered And Mutable(Changeable)
#Create List
marks = [78, 85, 90, 88]
print("Marks:", marks)
print("First element:", marks[0])
print("last element:", marks[-1])

#Add new value
marks.append(95)
print("After adding:", marks)

#Remove value
marks.remove(85)
print("After removing 85:", marks)

#sort value
marks.sort()
print("sorted:", marks)
#The values of List are sequential and also modified
```

```python
#DICTIONARIES
#Dictionary can store the Data in "Key:Value" pair
Student = {
    "Name": "Abhinadan",
    "Age": 24,
    "course": "Data science",
    "marks":[80,85,90]
}
print("Student info:", Student)
print("Name:", Student["Name"])
print("course:", Student["course"])

#Add new key-value pair
Student["grade"] = "A"
print("updated Dictionary:", Student)
#In dictionary the data structured are lavelled-- usefull for datasets
```

```python
#FOR LOOP
#For Loop can iterate in List or Range
marks = [80,85,90,95]
total = 0
for m in marks:
    total += m
average = total/len(marks)
print("Average Marks:", average)
#loop can added the value of marks in total and then find average
```

```python
#Functions
#A function groups code that is used repeatedly
def calculate_average(marks):
    total = sum(marks)
    avg = total/len(marks)
    return avg

marks = [80,85,90,95]
average = calculate_average(marks)
print("Average Marks (using function):", average)

#A function is define with def
```

```python
#NumPy
#NumPy are used for numeric calculations and arrays
import numpy as np
arr = np.array([1,2,3,4,5])
print("Array:", arr)
print("Mean:",np.mean(arr))
print("Standard deviation:", np.std(arr))
print("Sum:",np.sum(arr))
```

```python
#Pandas
#Pandas = DataFrames
import pandas as pd 
#create DataFrame
data = {
    "Name":["Abhinadan", "Ravi", "Sneha"],
    "Marks":[85,90,78],
    "city":["Delhi","Mumbai","Jaipur"]
}
df = pd.DataFrame(data)
print(df)
#Access column
print("\nMarks column:")
print(df["Marks"])
#Filter data
print("\nStudent with Marks > 80:")
print(df[df["Marks"]>80])
#In Pandas we process the data like Excel table 
```
